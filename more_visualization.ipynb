```python
import pickle as pk
import matplotlib.pyplot as plt
```

```python
with open("population_wisconsin.pk","rb") as f:
    pop = pk.load(f)
```

```python
data = []
for key in pop:
    data.append(pop[key])
```

```python
plt.hist(data,edgecolor='black')
plt.title("Histogram: Distribution of Population over Counties")
plt.xlabel("Population")
plt.ylabel("Number of Counties")
plt.savefig('pop_inbalance.pdf')  



```
